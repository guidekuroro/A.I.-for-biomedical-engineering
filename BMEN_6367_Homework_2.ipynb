```python
import cv2 as cv
import matplotlib.pyplot as plt
from scipy import ndimage

image_original = cv.imread('/content/drive/MyDrive/University of Texas at Dallas/BMEN 6367/Homework/brain_mri.jpeg') # read the image from google drive
fig, (ax1, ax2) = plt.subplots(1, 2) # subplot of images left and right
plt.gray()  # show the filtered result in grayscale
image_result = ndimage.percentile_filter(image_original, percentile=20, size=20) # perform percentile filter
ax1.imshow(image_original) # show image
ax2.imshow(image_result) 
plt.show() # show scale of image
```

```python
import cv2
import matplotlib.pyplot as plt
from scipy import ndimage

image_original = cv2.imread('/content/drive/MyDrive/University of Texas at Dallas/BMEN 6367/Homework/hand_x_ray.jpg',0) # read the image from google drive
fig, (ax1, ax2) = plt.subplots(1, 2) # subplot of images left and right
plt.gray()  # show the filtered result in grayscale
image_result = cv2.equalizeHist(image_original) # perform histogram equalization
ax1.imshow(image_original) # show image
ax2.imshow(image_result) 
plt.show() # show scale of image
```
